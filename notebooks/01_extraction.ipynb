```python
# =========================================
# 01_extraction.ipynb
# NYC Rolling Sales Dataset
# Purpose: Load raw dataset + understand data
# =========================================

# -----------------------------------------
# 1. Import Libraries
# -----------------------------------------
import pandas as pd
import numpy as np
```

```python
# -----------------------------------------
# 2. Load Raw Dataset
# -----------------------------------------
file_path = "../data/raw/nyc-rolling-sales.csv"
df = pd.read_csv(file_path)
print("Dataset Loaded Successfully ")
```

```python
# -----------------------------------------
# 3. Preview Dataset
# -----------------------------------------
print("\nFirst 5 Rows:")
display(df.head())

print("\nRandom Sample Rows:")
display(df.sample(5))
```

```python
# -----------------------------------------
# 4. Basic Dataset Information
# -----------------------------------------
print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Info:")
df.info()
```

```python
# -----------------------------------------
# 5. Statistical Summary
# -----------------------------------------
print("\nNumerical Summary:")
display(df.describe())

print("\nCategorical Summary:")
display(df.describe(include='object'))
```

```python
# -----------------------------------------
# 6. Missing Values Analysis
# Note: This dataset uses "-" for missing numeric values
# and blank strings for missing text values instead of NaN
# -----------------------------------------

# Step 1: Replace dash "-" and blank strings with actual NaN
df_check = df.copy()
for col in df_check.columns:
    df_check[col] = df_check[col].astype(str).str.strip()
    df_check[col] = df_check[col].replace(["-", "", "nan", "None", "NaN"], pd.NA)

# Step 2: Missing Values Count
print("Missing Values Count (dash and blank included as missing):")
missing_values = df_check.isnull().sum()
display(missing_values[missing_values > 0].sort_values(ascending=False))

# Step 3: Missing Values Percentage
print("Missing Values Percentage:")
missing_percentage = (df_check.isnull().sum() / len(df_check)) * 100
display(missing_percentage[missing_percentage > 0].sort_values(ascending=False).round(2))
```

```python
# -----------------------------------------
# 7. Duplicate Check
# -----------------------------------------
total_duplicates = df.duplicated().sum()
print("Duplicate Rows Count:", total_duplicates)
print(f"Total Duplicate Rows: {total_duplicates} out of {len(df)} rows")
```

```python
# -----------------------------------------
# 8. Unique Values Analysis
# -----------------------------------------
print("\nUnique Values Per Column:")
display(df.nunique().sort_values(ascending=False))
```

```python
# -----------------------------------------
# 9. Data Type Check
# -----------------------------------------
print("\nColumns with Object Data Type:")
object_columns = df.select_dtypes(include='object').columns.tolist()
print(object_columns)
```

```python
# -----------------------------------------
# 10. Memory Usage
# -----------------------------------------
print("\nMemory Usage (MB):")
memory_usage_mb = df.memory_usage(deep=True).sum() / 1024**2
print(round(memory_usage_mb, 2), "MB")
```

```python
# -----------------------------------------
# 11. Invalid Value Detection
# -----------------------------------------
if "SALE PRICE" in df.columns:
    zero_sale_price = (pd.to_numeric(df["SALE PRICE"], errors="coerce") == 0).sum()
    print("\nRows where SALE PRICE = 0:")
    print(zero_sale_price)

if "GROSS SQUARE FEET" in df.columns:
    zero_sqft = (pd.to_numeric(df["GROSS SQUARE FEET"], errors="coerce") == 0).sum()
    print("\nRows where GROSS SQUARE FEET = 0:")
    print(zero_sqft)
```

```python
# -----------------------------------------
# 12. Check Column Formatting Issues
# -----------------------------------------
print("\nRaw Column Names (Check for spaces/issues):")
for col in df.columns:
    print(f"'{col}'")
```

```python
# -----------------------------------------
# 13. Initial Business Variable Identification
# -----------------------------------------
important_columns = [
    "BOROUGH",
    "SALE PRICE",
    "GROSS SQUARE FEET",
    "BUILDING CLASS CATEGORY",
    "SALE DATE"
]

print("\nImportant Business Columns Present:")
for col in important_columns:
    if col in df.columns:
        print(f" {col}")
    else:
        print(f" {col}")
```

```python
# -----------------------------------------
# 15. Check for Blank Strings (Very Important)
# -----------------------------------------
# Sometimes values are not null, but just empty spaces.
print("\nBlank Strings Check:")
for col in df.select_dtypes(include='object').columns:
    blank_count = (df[col].astype(str).str.strip() == "").sum()
    if blank_count > 0:
        print(f"{col}: {blank_count} blank values")
```

```python
# -----------------------------------------
# 17. Check Negative Values
# -----------------------------------------
print("\nNegative Values Check:")
if "SALE PRICE" in df.columns:
    negative_prices = (pd.to_numeric(df["SALE PRICE"], errors="coerce") < 0).sum()
    print(f"Negative Sale Prices: {negative_prices}")
if "GROSS SQUARE FEET" in df.columns:
    negative_sqft = (pd.to_numeric(df["GROSS SQUARE FEET"], errors="coerce") < 0).sum()
    print(f"Negative Gross Square Feet: {negative_sqft}")
```

```python
# -----------------------------------------
# 18. Frequency Distribution of Key Categories
# -----------------------------------------
print("\nFrequency Distribution for BOROUGH:")
if "BOROUGH" in df.columns:
    display(df["BOROUGH"].value_counts())
```

```python
# -----------------------------------------
# 19. Date Range Check
# -----------------------------------------
print("\nDate Range Check:")
if "SALE DATE" in df.columns:
    temp_date = pd.to_datetime(df["SALE DATE"], errors="coerce")
    print(f"Min Date: {temp_date.min()}")
    print(f"Max Date: {temp_date.max()}")
```

```python
# -----------------------------------------
# 20. Detect Suspicious Extreme Values
# -----------------------------------------
print("\nSuspicious Extreme Values for SALE PRICE (Top 10):")
if "SALE PRICE" in df.columns:
    display(pd.to_numeric(df["SALE PRICE"], errors="coerce").sort_values(ascending=False).head(10))
```

```python
# -----------------------------------------
# 21. High Cardinality Columns
# -----------------------------------------
print("\nHigh Cardinality Columns (> 50% unique values):")
high_cardinality = df.nunique()[df.nunique() > len(df)*0.5]
display(high_cardinality)
```

```python
# -----------------------------------------
# 22. Final Extraction Summary
# -----------------------------------------
total_duplicates = df.duplicated().sum()
missing_cols = df.isnull().sum()
missing_count = (missing_cols > 0).sum()

print("Extraction Summary Completed")
print("="*45)
print(f"Total Rows          : {df.shape[0]}")
print(f"Total Columns       : {df.shape[1]}")
print(f"Duplicate Rows      : {total_duplicates}")
print(f"Columns with Nulls  : {missing_count}")
print("="*45)
print("""
Key findings:
- Dataset successfully loaded
- Missing values identified
- Duplicate check completed
- Invalid values (zero/negative) detected
- Blank strings and extra spaces checked
- Date ranges and extreme values validated
- High cardinality columns found
- Data type mismatches identified
""")
```
